# What Drives Urban Mode Choice?
## A Discrete Choice Analysis of Commuter Travel in Amsterdam

---

> **Abstract.** This study examines how commuters in Amsterdam choose between bus, metro, car, and bicycle. Using revealed preference data from the Dutch national travel survey ODiN 2023 combined with network-level travel time and frequency data derived from the national GTFS feed, a Multinomial Logit (MNL) model is estimated to quantify the marginal utility of travel time, generalised cost, access time, and waiting time. The model yields an empirical estimate of the Value of Travel Time (VoT) for Amsterdam commuters, which is benchmarked against the Dutch national reference value published by KiM (2023). The analysis further investigates the role of travel time variability on public transport uptake and the differential effect of peak-hour conditions across modes.

---

## 1. Problem Formulation and Theoretical Foundation

### 1.1 Research Question and Significance

Understanding why urban commuters choose one mode of transport over another is central to transport planning and infrastructure investment appraisal. Every consequential decision — a new metro line, a cycling infrastructure programme, a congestion pricing scheme — rests on assumptions about how travellers trade travel time against cost, and how sensitive that trade-off is to the quality of access to the network. Amsterdam provides a particularly instructive setting for this analysis: it combines a dense public transport network with the highest cycling mode share of any major European city, creating a genuine four-way competition between bus, metro, car, and bicycle that is rarely observable at this scale. This study addresses the following research question:

> *How do commuters in Amsterdam respond to travel time, cost, and access conditions when choosing between bus, metro, car, and bicycle — and what does this imply about the Value of Travel Time in this city?*

### 1.2 Random Utility Theory

The theoretical foundation of this study is Random Utility Theory (RUT), introduced in its modern discrete choice form by McFadden (1974). RUT posits that a decision-maker $n$ assigns a utility $U_{in}$ to each alternative $i$ in a finite choice set $C$, and selects the alternative that maximises that utility. Utility is decomposed into a systematic component $V_{in}$, which is a function of observed attributes, and a stochastic component $\varepsilon_{in}$, which captures unobserved factors and taste heterogeneity (Ben-Akiva and Lerman, 1985):

$$U_{in} = V_{in} + \varepsilon_{in}$$

The traveller chooses mode $i^*$ such that $U_{i^*n} \geq U_{jn}$ for all $j \in C$. Because $\varepsilon_{in}$ is unobserved by the analyst, the outcome is modelled probabilistically. The distribution assumed for the error term determines the functional form of the resulting choice model (Train, 2009).

### 1.3 The Multinomial Logit Model

When the error terms $\varepsilon_{in}$ are assumed to be independently and identically distributed (i.i.d.) according to the Extreme Value Type I (Gumbel) distribution, the resulting choice model is the Multinomial Logit (MNL), first derived by McFadden (1974). The systematic utility of mode $i$ for commuter $n$ is specified as a linear-in-parameters function of level-of-service attributes:

$$V_{in} = \text{ASC}_i + \beta_{tt}\, tt_{in} + \beta_{c}\, c_{in} + \beta_{acc}\, acc_{in} + \beta_{w}\, w_{in}$$

where $\text{ASC}_i$ is an Alternative Specific Constant capturing the average unobserved utility of mode $i$; $tt_{in}$ is in-vehicle travel time (minutes); $c_{in}$ is the generalised cost (€); $acc_{in}$ is access and egress time (minutes); and $w_{in}$ is waiting time (minutes), defined as zero for car and bicycle. One $\text{ASC}$ is normalised to zero for identification — in this study, the bicycle alternative serves as the reference category. The MNL choice probability is (McFadden, 1974; Train, 2009):

$$P(i \mid n) = \frac{\exp(V_{in})}{\sum_{j \in C} \exp(V_{jn})}$$

The model is estimated by Maximum Likelihood Estimation (MLE), maximising the log-likelihood over all observed choices. Parameters are estimated using PandasBiogeme (Bierlaire, 2023), the standard Python library for discrete choice model estimation.

### 1.4 Value of Travel Time

A primary output of the MNL model is the Value of Travel Time (VoT) — the monetary rate at which a traveller is willing to trade cost for a reduction in travel time. It is derived directly from the ratio of the marginal utility of time to the marginal utility of cost (Train, 2009):

$$VoT = -\frac{\beta_{tt}}{\beta_{c}} \times 60 \qquad [\text{€ per hour}]$$

The factor of 60 converts the per-minute ratio into an hourly figure, consistent with Dutch transport appraisal standards. Because VoT is a ratio of two estimated quantities, its variance is obtained via the Delta method (Train, 2009):

$$\text{Var}\!\left(\frac{\hat{\beta}_{tt}}{\hat{\beta}_{c}}\right) \approx \frac{1}{\hat{\beta}_{c}^{2}} \left[ \text{Var}(\hat{\beta}_{tt}) + \frac{\hat{\beta}_{tt}^{2}}{\hat{\beta}_{c}^{2}} \text{Var}(\hat{\beta}_{c}) - \frac{2\hat{\beta}_{tt}}{\hat{\beta}_{c}} \text{Cov}(\hat{\beta}_{tt}, \hat{\beta}_{c}) \right]$$

This yields a 95% confidence interval around the VoT estimate, enabling a formal comparison against the Dutch national benchmark.

### 1.5 Prior Literature and Benchmark Values

The MNL model remains the workhorse of applied mode choice analysis despite the availability of more flexible specifications such as the Nested Logit and Mixed Logit (Train, 2009; Ben-Akiva and Lerman, 1985). Its persistence reflects a favourable balance between parsimony, interpretability, and tractability, particularly when the primary objective is parameter estimation rather than individual-level prediction. Its principal limitation is the Independence of Irrelevant Alternatives (IIA) property, which imposes proportional substitution between all alternatives regardless of their similarity (McFadden, 1974). This is a relevant concern in the Amsterdam context, where bus and metro share access infrastructure and may exhibit correlated unobserved utilities. The IIA assumption is tested and relaxed in Section 4 through a Nested Logit specification that groups bus and metro as a correlated public transport nest.

For the Netherlands, KiM (2023) reports an official commuting VoT of €10.33 per hour, derived from large-scale stated and revealed preference studies commissioned by the Dutch Ministry of Infrastructure and Water Management. This figure serves as the primary external benchmark against which the estimated VoT is assessed in Section 5.

### 1.6 Hypotheses

Three testable hypotheses structure the empirical analysis. H1 and H2 are addressed through classical statistical testing in Section 3; H3 is addressed through MNL estimation and model validation in Section 4.

**H1 — Associative hypothesis.** There is a statistically significant negative association between travel time variability on public transport routes and the observed public transport mode share at the zonal level. More variable services are associated with lower PT uptake, though causality is not claimed given the presence of confounding variables such as income and residential density.

$$H_0: \rho\bigl(\bar{\sigma}_{\text{route}},\; s_{PT}\bigr) = 0 \qquad H_1: \rho\bigl(\bar{\sigma}_{\text{route}},\; s_{PT}\bigr) < 0$$

**H2 — Causal hypothesis.** Peak-hour conditions impose a significantly larger travel time variance penalty on car users than on metro users, reflecting the differential exposure of fixed-guideway versus road-based modes to congestion. Peak-hour status is treated as externally determined by clock time, supporting a causal interpretation subject to the caveat that route selection is endogenous.

$$H_0: \gamma_{\text{Mode} \times \text{Peak}} = 0 \qquad H_1: \gamma_{\text{Car} \times \text{Peak}} > \gamma_{\text{Metro} \times \text{Peak}}$$

**H3 — Choice behaviour hypothesis.** The MNL model yields statistically significant estimates for all level-of-service parameters, and the implied Value of Travel Time for Amsterdam commuters falls within the confidence interval of the Dutch national benchmark of €10.33 per hour (KiM, 2023). Furthermore, the Alternative Specific Constant for bicycle is significantly positive after controlling for travel time and cost, providing evidence of a genuine modal preference for cycling beyond what travel time and cost alone can explain.

$$H_0: \text{ASC}_{bike} = 0 \qquad H_1: \text{ASC}_{bike} > 0$$

### 1.7 Scope, Constraints, and Assumptions

The following constraints and assumptions apply throughout the study and are made explicit here in accordance with best practice in transport modelling (Ben-Akiva and Lerman, 1985; Train, 2009).

- **Geographic scope.** The analysis is restricted to commuting trips made by residents of the municipality of Amsterdam (CBS municipality code 363). Trips originating or ending outside Amsterdam are excluded.
- **Trip purpose.** The primary model is estimated on commuting trips only (ODiN travel purpose: *werken*). All-purpose trips are used for robustness checks in Section 4.
- **Choice set.** Four modes are considered: bus, metro or tram, car (driver), and bicycle. Modes with negligible mode share in Amsterdam (taxi, motorcycle, walking as a primary long-distance mode) are excluded from the choice set.
- **Revealed preference.** ODiN 2023 records actual choices, not stated preferences in hypothetical scenarios. This confers external validity but limits the ability to estimate willingness-to-pay for attributes not observed in the data, such as comfort or crowding.
- **Linearity of utility.** Systematic utility is specified as linear in parameters. Non-linear transformations of attributes (e.g., log of travel time) are considered as robustness checks.
- **IIA assumption.** The baseline MNL imposes proportional substitution between all modes. This is relaxed in the Nested Logit specification in Section 4.
- **Temporal scope.** ODiN 2023 is used as the primary data source. ODiN 2024 is excluded due to a confirmed methodological break identified by CBS (2024). Level-of-service attributes are derived from the Netherlands national GTFS feed, representative of scheduled services in 2023.
